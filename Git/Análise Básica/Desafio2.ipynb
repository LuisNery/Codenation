### Plotando as distribuições normais

```python
x_min = 350
x_max = 950

plt.figure(0,figsize=[12.4, 6.8])

x = np.linspace(x_min, x_max, 50)

yPR = st.norm.pdf(x,648.96,98.607)
yRS = st.norm.pdf(x,651.10,95.136) 
ySC = st.norm.pdf(x,649.54,97.233)

plt.plot(x,yPR, color='coral')
plt.plot(x,yRS, color='blue')
plt.plot(x,ySC, color='red')

plt.grid()

plt.xlim(x_min,x_max)

plt.title('Curvas normais de cada estado',fontsize=10)

plt.xlabel('Pontuação de crédito')

plt.legend(['PR','RS','SC'])
plt.show()
```
